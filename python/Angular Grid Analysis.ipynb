```python
import numpy as np
import matplotlib.pyplot as plt
import ipyvolume as ipv
%matplotlib inline
```

```python
x, w = np.polynomial.legendre.leggauss(20)
z = np.zeros_like(x)
```

```python
np.diff(np.arccos(x))
```

```python
theta = x*np.pi/2
plt.plot(z, x, 'o')
plt.plot(np.cos(theta), np.sin(theta))
for th in theta:
    plt.plot([0,np.cos(th)], [0, np.sin(th)])
plt.gca().set_aspect(1)
```

```python
phi_vals = (x+1)/2 * np.pi
theta_vals = (x+1)/2 * 2*np.pi
```

```python
th, ph = np.meshgrid(theta_vals, phi_vals, indexing='ij')
```

```python
x_sph = np.sin(ph) * np.cos(th)
y_sph = np.sin(ph) * np.sin(th)
z_sph = np.cos(ph)
```

```python
ipv.figure()
ipv.plot_surface(x_sph,y_sph,z_sph)
ipv.show()
```
