```python
def f(x,y):
    x0=1
    y0=2
    r = sqrt((x-x0)**2+(y-y0)**2)
    return cos(x+y)*exp(-r**2)+2 - x**2/100 -y**3/100

t = linspace(0,5,101)
y,x = meshgrid(t,t)
cont = contourf(x,y,f(x,y),cmap='YlGn')
title(r'$l_{n+1} = l^n + h_t G(\theta_f,l_n)$')
xlabel(r'$\theta$')
ylabel(r'$\l$')
cb = colorbar()
cb.ax.set_yticklabels([r'$l_{}$'.format(i) for i in range(9)]);


t = linspace(0,5,101)
y,x = meshgrid(t,t)
cont = contourf(x,y,f(x,y),colors='b',alpha=0.5,levels=(1.5,1.75))
cont = contour(x,y,f(x,y),levels=(1.5,1.75),colors='k')
title(r'$l_{n+1} = l^n + h_t G(\theta_f,l_n)$')
xlabel(r'$\theta$')
ylabel(r'$\l$')
cb = colorbar()
```
