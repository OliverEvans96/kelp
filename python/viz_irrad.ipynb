```python
import numpy as np
import ipyvolume as ipv
import h5py
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
irradfile = '../hdf5/kelp3d/irrad.hdf5'
kelpfile = '../hdf5/kelp3d/p_kelp.hdf5'
absfile = '../hdf5/kelp3d/abs_grid.hdf5'
radfile = '../hdf5/kelp3d/rad.hdf5'
with h5py.File(irradfile) as file:
    dset = file.require_dataset('irradiance', (10,10,10), 'd')
    irradiance = dset[...].T
with h5py.File(kelpfile) as file:
    dset = file.require_dataset('p_kelp', (10,10,10), 'd')
    p_kelp = dset[...].T
with h5py.File(absfile) as file:
    dset = file.require_dataset('p_kelp', (10,10,10), 'd')
    abs_grid = dset[...].T
with h5py.File(radfile) as file:
    dset = file.require_dataset('radiance', (10,10,10,10,10), 'd')
    radiance = dset[...].T
```
